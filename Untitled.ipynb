```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import matplotlib.pyplot as plt
import os
```

```python

df_dataset = pd.read_csv(r"C:\Users\songx\Desktop\学校\深度学习12个项目实践\深度学习12个项目实践\数据集\stock_1500.csv")

# 手动设置有意义的列名，这里假设数据有120列（从 'Unnamed: 0' 到 '118'）

column_names = ['index'] + [f'feature_{i}' for i in range(119)]
df_dataset.columns = column_names

# 假设数据集中 'feature_10' 是目标列（例如收盘价等），这里需要根据实际情况修改列名

target_column = 'feature_10'
features = df_dataset.drop(columns=[target_column])
target = df_dataset[target_column]

# 划分训练集和测试集
train_size = int(0.7 * len(df_dataset))
train_features = features[:train_size].values
train_target = target[:train_size].values
test_features = features[train_size:].values
test_target = target[train_size:].values

# 自定义数据集类
class StockDataset(Dataset):
    def __init__(self, features, target):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.target = torch.tensor(target, dtype=torch.float32)

    def __len__(self):
        return len(self.target)

    def __getitem__(self, idx):
        return self.features[idx], self.target[idx]

train_dataset = StockDataset(train_features, train_target)
test_dataset = StockDataset(test_features, test_target)

# 创建数据加载器
train_dataloader = DataLoader(train_dataset, batch_size=256, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=256)

# 选择设备
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# 构建适用于股票数据的简单卷积神经网络模型
class StockCNN(nn.Module):
    def __init__(self, input_size):
        super(StockCNN, self).__init__()
        self.conv1 = nn.Conv1d(1, 16, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.fc1 = nn.Linear(int(input_size / 2) * 16, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, x):
        x = x.unsqueeze(1)  # 增加通道维度
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# 根据特征数量初始化模型
input_size = train_features.shape[1]
model = StockCNN(input_size).to(device)

# 定义损失函数和优化器
loss_func = nn.MSELoss()  # 回归问题使用均方误差损失
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# 训练模型函数
def train_model(model, train_dataloader, optimizer, loss_func):
    total_loss = 0.
    model.train()
    for (x, y) in train_dataloader:
        x = x.to(device)
        y = y.to(device).unsqueeze(1)
        y_hat = model(x)
        loss = loss_func(y_hat, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    print(f"Total loss: {total_loss/len(train_dataloader):.4f}")
    model.eval()
    return total_loss/len(train_dataloader)

# 验证模型函数
def test_model(model, test_dataloader, loss_func):
    total_loss = 0.
    model.eval()
    for (x, y) in test_dataloader:
        x = x.to(device)
        y = y.to(device).unsqueeze(1)
        y_hat = model(x)
        loss = loss_func(y_hat, y)

        total_loss += loss.item()
    print(f"Total loss: {total_loss/len(test_dataloader):.4f}")
    model.train()
    return total_loss/len(test_dataloader)

epoch = 100
train_loss_records = []
test_loss_records = []

model = model.to(device)

for i in range(epoch):
    train_loss = train_model(model, train_dataloader, optimizer, loss_func)
    test_loss = test_model(model, test_dataloader, loss_func)
    train_loss_records.append(train_loss)
    test_loss_records.append(test_loss)
```

```python
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
plt.figure(figsize=(8, 4))
plt.plot(train_loss_records, label="train loss")
plt.plot(test_loss_records, label="test loss")
plt.title("Stock CNN Model Train Loss")
plt.legend()
plt.grid()
plt.show()
```
